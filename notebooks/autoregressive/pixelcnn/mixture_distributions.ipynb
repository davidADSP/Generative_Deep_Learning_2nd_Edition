```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

plt.figure(figsize=(15, 6), dpi=80)
# plt.axis('off')

# Plot between -10 and 10 with .001 steps.
x_axis = np.arange(0, 255, 0.001)
# Mean = 0, SD = 2.
plt.plot(x_axis, norm.pdf(x_axis, 50, 30), linewidth=5)
plt.plot(x_axis, norm.pdf(x_axis, 100, 50), linewidth=5)
plt.plot(x_axis, norm.pdf(x_axis, 200, 20), linewidth=5)
plt.show()
```
